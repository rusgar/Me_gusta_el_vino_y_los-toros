# Vamos a intentar salir no solo el color del vino sino su calidad

```python
from sklearn.linear_model import LogisticRegression       # logistic regression 
from sklearn.model_selection import train_test_split      # train/test data
from sklearn.linear_model import Ridge      

import math
from sklearn.metrics import accuracy_score, recall_score, precision_score, fbeta_score, classification_report
from sklearn.metrics import explained_variance_score, mean_absolute_error, r2_score, mean_squared_error
from sklearn.metrics import confusion_matrix
```

### Hay 11 funciones y 1 objetivo. SegÃºn esas caracterÃ­sticas y el objetivo, definamos las dos variables X como un conjunto de datos de caracterÃ­sticas "e" y como un conjunto de datos de destino. AsÃ­ que aquÃ­ la calidad de cada vino es nuestra variable objetivo

```python
df_wine
```

```python
#Definimos que valores queremos coger de la "X" e "y"; eliminamos quality pues ahora no la necesitamos
nombres= list(df_wine.columns)
nombres.remove('quality')

target = "quality"

X = df_wine[nombres]
y = df_wine[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Modelos

# [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) 

Escojemos este modelo por ser lineal porque hemos utilizado los minimos cuadrados ordinariospara minimizar la suma residual de cuadrados entre los objetivos observados en el conjunto de datos y los objetivos predichos por la aproximaciÃ³n lineal, lo veremos en **More_analytic_wine_red**

```python
lr = LogisticRegression(solver="liblinear")  
lr.fit(X_train, y_train)
```

```python
#Ajustamos el modelo
lr.fit(X_train, y_train)

#Sacamos la puntuacion del modelo de entrenamiento
lr.score(X_train, y_train)
```

```python
# Sacamos la puntuacion del test
print(lr.score(X_test, y_test))
```

```python
# Hacer predicciones contra el conjunto de prueba

preds = lr.predict(X_test)

variacion_test = explained_variance_score(y_test, preds)
mae = math.sqrt(mean_absolute_error(y_test, preds)) #PÃ©rdida de regresiÃ³n del error absoluto medio
rmse = math.sqrt(mean_squared_error(y_test, preds)) # PÃ©rdida de regresiÃ³n del error cuadrÃ¡tico medio.
r2 = math.sqrt(r2_score(y_test, preds)) # FunciÃ³n de puntuaciÃ³n de regresiÃ³n.
    
print("variacion_test = {:.4f} | MAE = {:.4f} | RMSE = {:.4f} | R2 = {:.4f}"
          .format(variacion_test, mae, rmse, r2))
```

```python
# Miramos la correlacion
df_wine_train.corr().loc["color"].abs().sort_values(ascending=False)
```
