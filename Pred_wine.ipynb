### Hay 11 funciones y 1 objetivo. Según esas características y el objetivo, defina las dos variables X como un conjunto de datos de características e y como un conjunto de datos de destino. Así que aquí la calidad es nuestra variable objetivo

```python
df_wine
```

```python
features = list(df_wine.columns)
features.remove('quality')

target = "quality"

X = df_wine[features]
y = df_wine[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

### Modelado con regresión logística

# [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

```python
reglog = LogisticRegression(solver="liblinear")  
reglog.fit(X_train, y_train)
```

```python
reglog.fit(X_train, y_train)

##Puntuacion del modelo de entrenamiento
reglog.score(X_train, y_train)    
```

```python
# Puntuacion del modelo de test
print(reglog.score(X_test, y_test))    
```

### Eliminar las variables altamente correlacionadas para mejorar la precisión media (puntuación)

```python
rev_features = list(df_wine.columns)
rev_features.remove('chlorides')
rev_features.remove('volatile acidity')
rev_features.remove('total sulfur dioxide')



target = "quality"

X_rev = df_wine[rev_features]
y_rev = df_wine[target]


X_rev_train, X_rev_test, y_rev_train, y_rev_test = train_test_split(X_rev, y_rev, test_size=0.25, random_state=0)  
```

```python
# Construya un nuevo modelo de regresión logística 
rev_reg_log = LogisticRegression(solver="liblinear")   
rev_reg_log.fit(X_rev_train, y_rev_train)

print(rev_reg_log.score(X_rev_train, y_rev_train), rev_reg_log.score(X_rev_test, y_rev_test))
```

### El puntaje de Train mejoró un poco, sin embargo, vemos una disminución en el puntaje de la prueba.

```python
X_preds = reglog.predict(X_test)
y_preds = reglog.predict(X_train)

variacion_test = explained_variance_score(y_test, X_preds)
variacion_train = explained_variance_score(y_rev_train, y_preds)

print("variacion_test = {:.4f} ~~~~~~~~ variacion_train = {:.4f}  "
          .format(variacion_test, variacion_train))
```

```python
# Hacer predicciones contra el conjunto de prueba

preds = reglog.predict(X_test)

variacion_test = explained_variance_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = math.sqrt(mean_squared_error(y_test, preds))
r2 = r2_score(y_test, preds)
    
print("variacion_test = {:.4f} | MAE = {:.4f} | RMSE = {:.4f} | R2 = {:.4f}"
          .format(variacion_test, mae, rmse, r2))
```
